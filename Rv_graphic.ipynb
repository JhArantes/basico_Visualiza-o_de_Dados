# Importando Pandas

```python
import pandas as pd
```

```python
# Salvando o dado em csv dentro de DF
df = pd.read_csv('imigrantes_canada.csv')
```

```python
# Lendo o DF, o que estiver na ultima linha sempre será mostrado 
df
```

```python
# O describe nos traz uma série de informações importantes em analise de dados

#        Count     ,  mean,    STD        , Min | 25% |50%/média| 75% | Max
# Numero de valores, média, Desvio Padrão e            os quartis
df.describe()
```

```python
# Aqui eu coloco os paises como index
df.set_index('País', inplace=True)

# Uso o MAP pra gerar uma lista porém os numéros viram strings ""STR""
#
anos = list(map(str, range(1980, 2014)))
```

```python
# Usando o loc para achar o index Brasil e pegar seus dados
# anos é quais colunas eu vou pegar 
brasil = df.loc['Brasil', anos]
brasil
```

# Criando DF    dados_brasil



```python
# Um Data frame puro nada mais é que um {dicionario : 1, 2, 3}

# Transformamos o ano que era um index em lista e 
#      colocamos o numero de imigrantes relativo a cada ano +++++++++++
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)

# Mostra os      3 primeiros  
dados_brasil.head(3)
```

# Importando Matplotlib
#### Porque usar o Matplotlib:
Ele nos permite ter uma maior controle na hora de plotar grafico 
além de ter sido criado por cima do pyplot que tem conecções com
importantes bibliotecas como : Pandas, Seaborn e etc..

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

## Adicionando e organizando estilo ao grafico 

```python
#tamanho da Grafico == Imagem
plt.figure(figsize=(8,4))

# Plot == Grafico de linha
# Eixo X e Y entra em plano cartezino  + info em: https://brasilescola.uol.com.br/o-que-e/matematica/o-que-e-plano-cartesiano.htm
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])

# Estilizando Grafico. Segue a ordem: titulo, Nome/rotulo de XeY, valores de x
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

# Agora vamos mecher com figuras no plt

```python
# Transformando em um figure que nos permite melhor manipulação e recursos
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])

# O Codigo automatiza pulando os anos de 5 em 5 no "axis X"|| xaxis
# ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

# Colocando titulo e legenda na figura
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_ylabel('Número de imigrantes')
ax.set_xlabel('Ano')

# Quebra o codigo e mostra apenas o grafico
plt.show()
```

# Criando mais de 1 grafico na Figure

Para isso passamos no subplots 2 parametros (qts de linhas, qts graphic por linha)

### Usando as cordenadas para se referir a cada graphic dentro da figura

```python
# Criando a figura      #QTD#
fig, axs = plt.subplots(1, 2, figsize=(15,5))

# Subplot 1: Gráfico de linhas
#  Cord
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))

# Subplot 2: Boxplot
#  Cord
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Box-plot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de imigrantes')

plt.show()
```

 Mais Exemplos 1+ Grafico dentro de uma figura

```python
# Criando a figura     #QTD#
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Ajustando o espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Subplots
axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')
#Cord
axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')
#Cord
axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')
#Cord
axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

# Alterando a frequência dos ticks do eixo X em todos os subplots com Loop For
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

plt.show()
```

## Finalizando estilização da figure

1. Espaçamento entre o graficos na figura 
2. Titulo da figure
3. Colocando em todos os graficos o set_major_locator(plt.MultipleLocator())     AXIS  X
4. Adicionando label em todos
5. Definindo o min e max do   AXIS  Y

```python
# Criando a figura
fig, axs = plt.subplots(2, 2, figsize=(10, 6))


# Ajustando o espaçamento
fig.subplots_adjust(hspace=0.5, wspace=0.3)
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

# Adicionando um título geral para os subplots
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

# Subplots
axs[0, 0].plot(df.loc['Brasil', anos])
axs[0, 0].set_title('Brasil')

axs[0, 1].plot(df.loc['Colômbia', anos])
axs[0, 1].set_title('Colômbia')

axs[1, 0].plot(df.loc['Argentina', anos])
axs[1, 0].set_title('Argentina')

axs[1, 1].plot(df.loc['Peru', anos])
axs[1, 1].set_title('Peru')

# Alterando a frequência dos ticks do eixo X em todos os subplots
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

# Adicionando rótulos para os eixos X e Y
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

# Definindo a mesma escala no eixo Y em todos os subplots
ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)
#=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-

plt.show()
```
